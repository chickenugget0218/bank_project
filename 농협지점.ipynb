```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
##다시
driver= webdriver.Chrome("chromedriver")
driver.get('https://nonghyup.ttmap.co.kr/main.jsp')
driver.implicitly_wait(3)

    # 농협클릭
driver.find_element(By.XPATH,'//*[@id="section_branch"]/ul/li[3]/label/span/a').click()

        # 검색버튼 클릭
driver.find_element(By.XPATH,'//*[@id="search_branchname"]/ul/li/input[3]').click()
driver.implicitly_wait(2)

            # 전체
driver.find_element(By.XPATH,'//*[@id="search_result"]/div[1]/div[2]/ul/li[1]/a').click()
time.sleep(2)

### 페이지 크롤링
## 페이지 1~599
curPage = 1
totalPage = 600

# 새로 저장할 리스트
new_title =[]
new_add =[]
new_tel=[]

while curPage < totalPage:

    #지점별 크롤링
    title=[]
    titles = driver.find_elements_by_css_selector("h2.list_title.nonghyup")
    for i in titles:
        title.append(i.text)
        # print(title) #들어가는지 확인      
    #저장
    new_title += title 
    #print(new_title)
    
    
     ## 주소
    address =[]
    adds = driver.find_elements_by_css_selector("p.list_addr")
    for i in adds:
        address.append(i.text)
        # print(address)
    #저장
    new_add += address 
    #print(new_add)


      ## 전화번호
    tel = []
    tels = driver.find_elements_by_css_selector("p.list_telno")
    for i in tels:
        tel.append(i.text)
        # print(tel)
    #저장
    new_tel += tel
    #print(new_tel)

    
    # 현재페이지 표시 후 다음페이지
    print("-------------------------current page is {0}--------------------------".format(curPage))
    curPage +=1
    
    driver.execute_script("goPage({0})".format(curPage))
    time.sleep(2)
```

```python
#마지막 원소 확인
print(new_title[-1])
```

```python
## df저장
import pandas as pd

df = pd.DataFrame({'branch':new_title,
                   'address':new_add,
                   'tel':new_tel})
```
